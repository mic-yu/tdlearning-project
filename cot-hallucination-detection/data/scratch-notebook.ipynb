```python
import json, pandas as pd
pd.set_option("display.max_colwidth", None)

# Load processed file
df = pd.read_csv("qwen-3b-instruct-gsm8k-responses-processed.csv")

# Peek at a random subset
for idx, row in df.sample(n=5, random_state=42).iterrows():
    print(f"\nRow {idx} | label={row['label']} | steps_parsed={row['steps_parsed']}")
    steps = json.loads(row["reasoning_steps_json"])
    for i, step in enumerate(steps, 1):
        print(f"\n  Trace {i}:\n{step}")
```

```python
import pandas as pd
from sklearn.model_selection import train_test_split

# Load processed data (same folder as the notebook)
df = pd.read_csv("qwen-3b-instruct-gsm8k-responses-processed.csv")

# Optional: keep only rows with parsed steps
df = df[df["steps_parsed"]]

# Split
train_df, test_df = train_test_split(df, test_size=0.2, random_state=42, shuffle=True)

# Write back to the same folder
train_df.to_csv("qwen-train.csv", index=False)
test_df.to_csv("qwen-test.csv", index=False)

print(len(train_df), "train rows,", len(test_df), "test rows")
```
